```python
# =========================================================
# Pipeline de geração de dataset sintético de concentração de ozônio
# =========================================================

import os
import sys
import shutil
from itertools import combinations
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.interpolate import griddata
from scipy.ndimage import gaussian_filter, shift
from skimage.transform import resize
import ipywidgets as widgets
from IPython.display import display

# =========================================================
# Constantes
# =========================================================

GRID_SIZE = 224
ALPHAS = [0.2, 0.4, 0.6, 0.8]
GAUSSIAN_STD = 0.02
SALT_PEPPER_PROB = 0.01
BLUR_SIGMA = 1.5
MAX_SHIFT = 5
DOWNSAMPLE_FACTOR = 2

# =========================================================
# Detectar ambiente
# =========================================================

def is_colab():
    return "google.colab" in sys.modules

# =========================================================
# Configuração automática do ambiente
# =========================================================

def configure_environment():
    print("\n==============================")
    print("Configuração do Ambiente")
    print("==============================")

    if is_colab():
        print("Ambiente detectado: Google Colab")
        from google.colab import drive
        drive.mount("/content/drive", force_remount=True)
        base_path = "/content/drive/MyDrive/Iniciacao_cientifica/Tarefa_1/ozone_source_localization_v2/data"
    else:
        print("Ambiente detectado: Execução local")
        base_path = "./sim/sim"

    if not os.path.exists(base_path):
        raise Exception(f"Caminho não encontrado: {base_path}")

    processed_path = os.path.join(base_path, "processed")
    augm_path = os.path.join(base_path, "augm")
    os.makedirs(augm_path, exist_ok=True)

    print(f"[INFO] Pasta processed: {processed_path}")
    print(f"[INFO] Pasta augm: {augm_path}")
    return processed_path, augm_path

# =========================================================
# Limpeza da pasta augm
# =========================================================

def ask_clean_directory(path):
    print("\n==============================")
    print("Verificação da pasta de saída")
    print("==============================")

    files = os.listdir(path)
    print(f"Pasta: {path}")
    print(f"Total de arquivos encontrados: {len(files)}")

    if len(files) == 0:
        print("Pasta já está vazia.\n")
        return

    answer = input("\nDeseja apagar os arquivos antigos antes de rodar? (s/n): ").lower()
    if answer == "s":
        print("\nRemovendo arquivos...")
        for file in files:
            file_path = os.path.join(path, file)
            try:
                if os.path.isfile(file_path):
                    os.remove(file_path)
                else:
                    shutil.rmtree(file_path)
            except Exception as e:
                print("Erro ao remover:", file, e)
        print("Pasta limpa.\n")
    else:
        print("Arquivos antigos serão mantidos.\n")

# =========================================================
# Carregar arquivos processados
# =========================================================

def list_processed_files(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Pasta 'processed' não encontrada:\n{path}")
    files = sorted([f for f in os.listdir(path) if f.endswith(".csv")])
    print(f"\nArquivos encontrados: {len(files)}")
    return files

def load_simulations(files, path):
    data = []
    names = []
    for f in files:
        df = pd.read_csv(os.path.join(path, f))
        data.append(df)
        names.append(f)
    return data, names

# =========================================================
# Interpolação
# =========================================================

def interpolate_field(df, size=GRID_SIZE):
    x, y, z = df["x"].values, df["y"].values, df["value"].values
    xi = np.linspace(x.min(), x.max(), size)
    yi = np.linspace(y.min(), y.max(), size)
    xi, yi = np.meshgrid(xi, yi)
    zi = griddata((x, y), z, (xi, yi), method="linear")
    return np.nan_to_num(zi, nan=0.0)

def convert_all_fields(data):
    grids = []
    for df in tqdm(data, desc="Interpolando campos"):
        grids.append(interpolate_field(df))
    return grids

# =========================================================
# Combinações lineares
# =========================================================

def combine_fields(field1, field2, alpha):
    return alpha * field1 + (1 - alpha) * field2

def generate_combinations(grids, names, output_path):
    pairs = list(combinations(range(len(grids)), 2))
    total = len(pairs) * len(ALPHAS)
    print("\nTotal de pares:", len(pairs))
    print("Total de campos sintéticos:", total)
    with tqdm(total=total, desc="Gerando campos") as pbar:
        for i, j in pairs:
            for alpha in ALPHAS:
                field = combine_fields(grids[i], grids[j], alpha)
                name1, name2 = names[i].replace(".csv",""), names[j].replace(".csv","")
                filename = f"comb_{name1}_{name2}_a{alpha:.1f}.npy"
                np.save(os.path.join(output_path, filename), field)
                pbar.update(1)

# =========================================================
# Data augmentation
# =========================================================

def add_gaussian_noise(field):
    return np.clip(field + np.random.normal(0, GAUSSIAN_STD, field.shape), 0, 1)

def add_salt_pepper(field):
    noisy = field.copy()
    salt = np.random.rand(*field.shape) < SALT_PEPPER_PROB
    pepper = np.random.rand(*field.shape) < SALT_PEPPER_PROB
    noisy[salt] = 1
    noisy[pepper] = 0
    return noisy

def blur_field(field):
    return gaussian_filter(field, sigma=BLUR_SIGMA)

def random_shift(field):
    dx, dy = np.random.uniform(-MAX_SHIFT, MAX_SHIFT), np.random.uniform(-MAX_SHIFT, MAX_SHIFT)
    return shift(field, (dx, dy), mode="nearest")

def downsample(field):
    small = resize(field, (field.shape[0] // DOWNSAMPLE_FACTOR, field.shape[1] // DOWNSAMPLE_FACTOR), anti_aliasing=True)
    return resize(small, field.shape, anti_aliasing=True)

def augment_field(field):
    aug, applied = field.copy(), []
    if np.random.rand() < 0.5: aug, applied = add_gaussian_noise(aug), applied + ["gauss"]
    if np.random.rand() < 0.3: aug, applied = add_salt_pepper(aug), applied + ["saltpepper"]
    if np.random.rand() < 0.4: aug, applied = blur_field(aug), applied + ["blur"]
    if np.random.rand() < 0.4: aug, applied = random_shift(aug), applied + ["shift"]
    if np.random.rand() < 0.3: aug, applied = downsample(aug), applied + ["downsample"]
    if len(applied) == 0: applied.append("none")
    return np.clip(aug, 0, 1), "_".join(applied)

def apply_augmentation(path):
    files = [f for f in os.listdir(path) if f.endswith(".npy")]
    print("\nCampos base:", len(files))
    for file in tqdm(files, desc="Aplicando augmentation"):
        field = np.load(os.path.join(path, file))
        augmented, aug_type = augment_field(field)
        new_name = file.replace(".npy", f"_{aug_type}.npy")
        np.save(os.path.join(path, new_name), augmented)

# =========================================================
# Visualização interativa
# =========================================================

def interactive_viewer(path):
    files = sorted([f for f in os.listdir(path) if f.endswith(".npy")])
    print("\nTotal de campos:", len(files))
    index = 0
    plot_area = widgets.Output()

    btn_prev = widgets.Button(description="⬅ Anterior")
    btn_next = widgets.Button(description="Próximo ➡", button_style="success")
    buttons = widgets.HBox([btn_prev, btn_next], layout=widgets.Layout(justify_content="center"))

    search_box = widgets.Text(placeholder="Digite parte do nome...")
    search_btn = widgets.Button(description="Ir", button_style="info")
    search_area = widgets.HBox([search_box, search_btn], layout=widgets.Layout(justify_content="center"))

    def show():
        with plot_area:
            plot_area.clear_output(wait=True)
            file = files[index]
            field = np.load(os.path.join(path, file))
            plt.figure(figsize=(6,6))
            im = plt.imshow(field, cmap="viridis")
            cbar = plt.colorbar(im, fraction=0.046, pad=0.04)
            cbar.set_label("Concentração de O3", rotation=270, labelpad=15)
            plt.title(f" {file}", fontsize=12, pad=25)
            plt.xlabel("X")
            plt.ylabel("Y")
            plt.show()

    def next_field(b):
        nonlocal index
        index = (index+1) % len(files)
        show()

    def prev_field(b):
        nonlocal index
        index = (index-1) % len(files)
        show()

    def search(b):
        nonlocal index
        text = search_box.value.lower()
        for i, f in enumerate(files):
            if text in f.lower():
                index = i
                show()
                return
        print("Arquivo não encontrado")

    btn_next.on_click(next_field)
    btn_prev.on_click(prev_field)
    search_btn.on_click(search)

    display(plot_area)
    display(buttons)
    display(search_area)
    show()

# =========================================================
# MAIN
# =========================================================

def main():
    processed_path, augm_path = configure_environment()
    ask_clean_directory(augm_path)
    files = list_processed_files(processed_path)
    data, names = load_simulations(files, processed_path)
    grids = convert_all_fields(data)
    generate_combinations(grids, names, augm_path)
    apply_augmentation(augm_path)
    interactive_viewer(augm_path)
    print("\nPipeline finalizado com sucesso!")

# =========================================================
# Execução
# =========================================================

if __name__ == "__main__":
    main()
```
